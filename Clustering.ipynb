```python
import csv
import pandas as pd
import numpy as np
df = pd.read_csv('3VAC1.csv')
df
```

```python
for i in range(len(df)):
    df['amplitudes'][i]= df['amplitudes'][i][1:len(df['amplitudes'][i])-2]
    df['amplitudes'][i] = np.array(df['amplitudes'][i].split(','))
    df['amplitudes'][i] =[float(i)/100 for i in df['amplitudes'][i]]
df
```

```python
am = df['amplitudes']
am = am.apply(pd.Series)
am
```

```python
f = [df['device_id'],am]
data = pd.concat(f,axis=1)
data
```

```python
from sklearn.preprocessing import StandardScaler

# Separating out the features
x = am.values
# Separating out the target
y = data['device_id'].values
# Standardizing the features
x = StandardScaler().fit_transform(x)
x
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(am)
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['principal component 1', 'principal component 2'])
principalDf
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 component PCA', fontsize = 20)



ax.scatter(principalDf['principal component 1']
               , principalDf['principal component 2'])

ax.grid()
```

```python
pca.explained_variance_ratio_
```

## Spectral Clustering

```python
from sklearn.cluster import SpectralClustering
import numpy as np
clustering = SpectralClustering(n_clusters=6,assign_labels='discretize',random_state=0).fit(am)
labels = clustering.labels_
```

```python
cluster1 = pd.concat([df['device_id'],pd.DataFrame(labels),am],axis=1)
cluster1
```

## kmeans


```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=6)
kmeans.fit(am)
y_kmeans = kmeans.predict(am)
y_kmeans
```

```python
cluster2 = pd.concat([df['device_id'],pd.DataFrame(y_kmeans),am],axis=1)
cluster2
```

```python
Sum_of_squared_distances = []
K = range(1,15)
for k in K:
    km = KMeans(n_clusters=k)
    km = km.fit(am)
    Sum_of_squared_distances.append(km.inertia_)
```

```python
plt.plot(K, Sum_of_squared_distances, 'bx-')
plt.xlabel('k')
plt.ylabel('Sum_of_squared_distances')
plt.title('Elbow Method For Optimal k')
plt.show()
```

## Affinity Propagation

```python
from sklearn.cluster import AffinityPropagation
from matplotlib import pyplot
# define the model
model = AffinityPropagation(damping=0.9)
# fit the model
model.fit(am)
# assign a cluster to each example
yhat = model.predict(am)
yhat
```

```python
cluster3 = pd.concat([df['device_id'],pd.DataFrame(yhat),am],axis=1)
cluster3
```

## Agglomerative Clustering

```python
from sklearn.cluster import AgglomerativeClustering
# define the model
model = AgglomerativeClustering(n_clusters=6)
# fit model and predict clusters
yhat2 = model.fit_predict(am)
```

```python
cluster4 = pd.concat([df['device_id'],pd.DataFrame(yhat2),am],axis=1)
cluster4
```

## GaussianMixture

```python
from sklearn.mixture import GaussianMixture
# define the model
model = GaussianMixture(n_components=4)
# fit model and predict clusters
yhat3 = model.fit_predict(am)
```

```python
cluster5 = pd.concat([df['device_id'],pd.DataFrame(yhat3),am],axis=1)
cluster5
```
