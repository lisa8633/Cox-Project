```python
import csv
import pandas as pd
import numpy as np
df = pd.read_csv('3VAC1.csv')
df
```

```python
for i in range(len(df)):
    df['amplitudes'][i]= df['amplitudes'][i][1:len(df['amplitudes'][i])-2]
    df['amplitudes'][i] = np.array(df['amplitudes'][i].split(','))
    df['amplitudes'][i] =[float(i)/100 for i in df['amplitudes'][i]]
df
```

```python
am = df['amplitudes']
am = am.apply(pd.Series)
am
```

```python
f = [df['device_id'],am]
data = pd.concat(f,axis=1)
data
```

```python
from sklearn.preprocessing import StandardScaler

# Separating out the features
x = am.values
# Separating out the target
y = data['device_id'].values
# Standardizing the features
x = StandardScaler().fit_transform(x)
x
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(am)
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['principal component 1', 'principal component 2'])
principalDf
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 component PCA', fontsize = 20)



ax.scatter(principalDf['principal component 1']
               , principalDf['principal component 2'])

ax.grid()
```

```python
pca.explained_variance_ratio_
```

## Spectral Clustering

```python
from sklearn.cluster import SpectralClustering
import numpy as np
clustering = SpectralClustering(n_clusters=6,assign_labels='discretize',random_state=0).fit(am)
labels = clustering.labels_
from sklearn.metrics import silhouette_score
spectral= silhouette_score(am, clustering.fit_predict(am))
print("Silhouette score:")
print(spectral)
from sklearn.metrics import calinski_harabasz_score
ch_spectral = calinski_harabasz_score(am, clustering.fit_predict(am))
print("Calinski Harabaz Index:")
print(ch_spectral)
```

```python
cluster1 = pd.concat([df['device_id'],pd.DataFrame(labels),am],axis=1)
cluster1
```

## kmeans


```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=6)
kmeans.fit(am)
y_kmeans = kmeans.predict(am)
from sklearn.metrics import silhouette_score
km= silhouette_score(am, kmeans.predict(am))
print("Silhouette score:")
print(km)
from sklearn.metrics import calinski_harabasz_score
ch_km = calinski_harabasz_score(am, kmeans.predict(am))
print("Calinski Harabaz Index:")
print(ch_km)
```

```python
cluster2 = pd.concat([df['device_id'],pd.DataFrame(y_kmeans),am],axis=1)
cluster2
```

```python
Sum_of_squared_distances = []
K = range(1,15)
for k in K:
    km = KMeans(n_clusters=k)
    km = km.fit(am)
    Sum_of_squared_distances.append(km.inertia_)
```

```python
plt.plot(K, Sum_of_squared_distances, 'bx-')
plt.xlabel('k')
plt.ylabel('Sum_of_squared_distances')
plt.title('Elbow Method')
plt.show()
```

## Affinity Propagation

```python
from sklearn.cluster import AffinityPropagation
from matplotlib import pyplot
# define the model
model = AffinityPropagation(damping=0.9)
# fit the model
model.fit(am)
# assign a cluster to each example
yhat = model.predict(am)
from sklearn.metrics import silhouette_score
affinity= silhouette_score(am, model.predict(am))
print("Silhouette score:")
print(affinity)
from sklearn.metrics import calinski_harabasz_score
ch_affinity = calinski_harabasz_score(am, model.predict(am))
print("Calinski Harabaz Index:")
print(ch_affinity)
```

```python
cluster3 = pd.concat([df['device_id'],pd.DataFrame(yhat),am],axis=1)
cluster3
```

## Hiearchical

```python
import scipy.cluster.hierarchy as shc
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=(10, 7))  
plt.title("Ward Dendrograms")  
dend = shc.dendrogram(shc.linkage(am, method='ward'))
plt.axhline(y=5300, color='r', linestyle='--')
#plt.xticks(rotation = 45)
```

```python
plt.figure(figsize=(10, 7))  
plt.title(" Complete Dendrograms")  
dend = shc.dendrogram(shc.linkage(am, method='complete'))
plt.axhline(y=1250, color='r', linestyle='--')
#plt.xticks(rotation = 45)
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=7, affinity='euclidean', linkage='ward')  
d = cluster.fit_predict(am)
from sklearn.metrics import silhouette_score
hiearch = silhouette_score(am, cluster.fit_predict(am))
print("Silhouette score:")
print(hiearch)
from sklearn.metrics import calinski_harabasz_score
ch_hiearch = calinski_harabasz_score(am, cluster.fit_predict(am))
print("Calinski Harabaz Index:")
print(ch_hiearch)
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=6, affinity='euclidean', linkage='complete')  
d = cluster.fit_predict(am)
from sklearn.metrics import silhouette_score
hiearch2 = silhouette_score(am, cluster.fit_predict(am))
print("Silhouette score:")
print(hiearch2)
from sklearn.metrics import calinski_harabasz_score
ch_hiearch2 = calinski_harabasz_score(am, cluster.fit_predict(am))
print("Calinski Harabaz Index:")
print(ch_hiearch2)
```

```python
cluster5 = pd.concat([df['device_id'],pd.DataFrame(d),am],axis=1)
cluster5
```

```python
#The Silhouette Score attempts to describe how similar a datapoint is to other datapoints in its cluster, 
#relative to datapoints not in its cluster 
#(this is aggregated over all datapoints to get the score for an overall clustering)
```

```python
m = ['Spectral','Affinity','Ward Linkage','Complete Linkage']
silh = [spectral,affinity,hiearch,hiearch2]

print(m)
print(silh)
```

```python
import matplotlib.pyplot as plt
   
plt.barh(m,silh)
plt.title('Silhouette score')
plt.ylabel('Models')
plt.xlabel('Silhouette score')
plt.show()
```

```python
#The Calinski Harabaz Index is the ratio of the variance of a datapoint compared to points in other clusters, 
#against the variance compared to points within its cluster.
```

```python
ch = [ch_spectral,ch_affinity,ch_hiearch,ch_hiearch2]

print(m)
print(ch)
```

```python
plt.barh(m,ch)
plt.title('Calinski Harabaz Index')
plt.ylabel('Models')
plt.xlabel('Calinski Harabaz Index')
plt.show()
```
