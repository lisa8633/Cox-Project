```python
import csv
import pandas as pd
import numpy as np
import os
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score
from sklearn.metrics import calinski_harabasz_score

path = './fbc'

files = os.listdir(path)
li = []
for f in files:
    li.append(f)
li.remove('.DS_Store')
li.remove('.ipynb_checkpoints')
li
```

```python
sh = []
ch = []
path = path = './fbc/'
for f in li:
    i1 = path + f
    df = pd.read_csv(i1)
    for i in range(len(df)):
        df['amplitudes'][i]= df['amplitudes'][i][1:len(df['amplitudes'][i])-2]
        df['amplitudes'][i] = np.array(df['amplitudes'][i].split(','))
        df['amplitudes'][i] =[float(i)/100 for i in df['amplitudes'][i]]
    am = df['amplitudes']
    am = am.apply(pd.Series)
    
    cluster = AgglomerativeClustering(n_clusters=9, affinity='euclidean', linkage='ward')  
    d = cluster.fit_predict(am)
    hiearch = silhouette_score(am, cluster.fit_predict(am))
    sh.append(hiearch)
    ch_hiearch = calinski_harabasz_score(am, cluster.fit_predict(am))
    ch.append(ch_hiearch)

```

```python
np.mean(ch)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(16,8))

plt.bar(li,sh)
plt.title('Silhouette score')
plt.xlabel('Nodes')
plt.ylabel('Silhouette score')
plt.xticks(rotation = 45)
plt.axhline(y=np.mean(sh), color='r', linestyle='--')
plt.show()
```

```python
plt.figure(figsize=(16,8))
plt.bar(li,ch)
plt.title('Calinski Harabaz Index')
plt.xlabel('Nodes')
plt.ylabel('Calinski Harabaz Index')
plt.xticks(rotation = 45)
plt.axhline(y=np.mean(ch), color='r', linestyle='--')
plt.show()
```

```python
from sklearn.cluster import AffinityPropagation
```

```python
ash = []
ach = []
num = []
path = path = './fbc/'
for f in li:
    i1 = path + f
    df = pd.read_csv(i1)
    for i in range(len(df)):
        df['amplitudes'][i]= df['amplitudes'][i][1:len(df['amplitudes'][i])-2]
        df['amplitudes'][i] = np.array(df['amplitudes'][i].split(','))
        df['amplitudes'][i] =[float(i)/100 for i in df['amplitudes'][i]]
    am = df['amplitudes']
    am = am.apply(pd.Series)
    
    model = AffinityPropagation(damping=0.9) 
    model.fit(am)
    cluster_centers_indices = model.cluster_centers_indices_

    no_clusters = len(cluster_centers_indices)
    num.append(no_clusters)
    d = model.predict(am)
    affs = silhouette_score(am, d)
    ash.append(affs)
    ch_aff = calinski_harabasz_score(am, d)
    ach.append(ch_aff)
```

```python
plt.figure(figsize=(16,8))

plt.bar(li,ash)
plt.title('Silhouette score')
plt.xlabel('Nodes')
plt.ylabel('Silhouette score')
plt.xticks(rotation = 45)
plt.axhline(y=np.mean(ash), color='r', linestyle='--')
plt.show()
```

```python
np.mean(ach)
```

```python
plt.figure(figsize=(16,8))
plt.bar(li,ach)
plt.title('Calinski Harabaz Index')
plt.xlabel('Nodes')
plt.ylabel('Calinski Harabaz Index')
plt.xticks(rotation = 45)
plt.axhline(y=np.mean(ach), color='r', linestyle='--')
plt.show()
```

```python
plt.figure(figsize=(16,8))
plt.bar(li,num)
plt.title('# of Clusters')
plt.xlabel('Nodes')
plt.ylabel('# of Clusters')
plt.xticks(rotation = 45)
plt.axhline(y=np.mean(num), color='r', linestyle='--')
plt.show()
```
